```python

def checkCousins(root, p, q):

	if root == None:
		return False

	
	parA = None

	
	parB = None

	
	queue = []

	
	tmp = Node(-1)

	
	queue.append((root, tmp))
	
	while len(q) > 0:

		levSize = len(queue)
		while levSize:

			ele = queue.pop(0)


			if ele[0].data == p.data:
				parA = ele[1]
			
			if ele[0].data == q.data:
				parB = ele[1]


			if ele[0].left:
				q.append((ele[0].left, ele[0]))
			
			if ele[0].right:
				q.append((ele[0].right, ele[0]))
			levSize -= 1

			
			if parA and parB:
				break

		
		if parA and parB:
			return parA != parB

		
		if (parA and not parB) or (parB and not parA):
			return False
		
	return False


```
